```python
import pandas as pd
jsonObj = pd.read_json(path_or_buf='output_fixed.jsonl', lines=True)
```

```python
jsonObj
```

```python
formatted_json_objects = []
system_message = "Marv is a factual chatbot offering information about San José State University's Masters in Data Analytics program, focusing on department-related website queries."


for i in range(len(jsonObj["prompt"])):
    formatted_json_objects.append({
        "messages": [
            {"role": "system", "content": system_message},
            {"role": "user", "content": jsonObj["prompt"][i]},
            {"role": "assistant", "content": jsonObj["completion"][i]}
        ]
    })
```

```python
formatted_json_objects
```

```python
import json
import random


jsonl_content = "\n".join(json.dumps(obj) for obj in random.sample(formatted_json_objects, 50))
```

```python
with open('data.jsonl', 'w', encoding='utf-8') as f:
    f.write(jsonl_content)
```
